```python
import csv
import requests
from bs4 import BeautifulSoup
import urllib3
from selenium import webdriver
import time
import json
```

```python
output = []
start_time = time.time()
with open("scrape.csv",'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    next(csv_reader)
    count = 0
    for line in csv_reader:
        dic = {}
        asin = line[2]
        country = line[3]
        url = f"https://www.amazon.{country}/dp/{asin}"
        # url = "https://www.amazon.de/dp/000004458X"
        # url = "https://www.amazon.de/dp/000104348X"
        # url = "https://www.amazon.de/dp/000101742X"
        # url = "https://www.amazon.fr/dp/000101742X"
        if count != 0 and count % 100 == 0:
            end_time = time.time()
            print(f"Time taken to complete 100 URLs is {end_time-start_time}")
            start_time = time.time()
        http = urllib3.PoolManager()
        error = http.request("GET", url)
        if error.status == 404 or error.status == 500:
            print(f"{url} not available")
            count += 1
            continue
        driver = webdriver.Chrome()
        driver.get(url)
        driver.implicitly_wait(20)
        try:
            product_title = driver.find_element_by_xpath('//h1/span').text
            dic["Product title"] = product_title
            print(product_title)
            dic["Image Url"] = None
            try:
                image_url = driver.find_element_by_xpath("//span/span/div/img")
                dic["Image Url"] = image_url.get_attribute('src')
                print(image_url.get_attribute('src'))
            except:
                image_url = driver.find_element_by_xpath("//div[1]/div/div/div/div/div/div/img")
                dic["Image Url"] = image_url.get_attribute("src")
                print(image_url.get_attribute('src'))
            dic["Product Information"] = None
            dic["Price"] = None
            try:
                prod_inf1 = driver.find_element_by_id("detailBullets_feature_div")
                dic["Product Information"] = prod_inf1.text
                print(prod_inf1.text)
            except:
                prod_inf1 = driver.find_element_by_id("feature-bullets")
                dic["Product Information"] = prod_inf1.text
                print(prod_inf1.text)
            try:
                price = driver.find_element_by_xpath("//span/a/span/span")
                dic["Price"] = price.text
                print(price.text)
            except:
                price = driver.find_element_by_class_name("apexPriceToPay")
                dic["Price"] = price.text
                print(price.text)
            count += 1
            output.append(dic)
            driver.quit()
        except:
            if dic != {}:
                output.append(dic)
            driver.quit()
            count += 1
            continue
        
```

```python
print(output)
```

```python
with open("output.json","w") as result:
    json.dumps(output,result)
```
